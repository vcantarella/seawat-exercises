# Elder's Case

- Case from Oude Essink's Exercise see Readme
- Model implemented in Flopy (Bakker et al. 2016)

## Step 1. Import libraries

```python
import flopy
import numpy as np
import os
import matplotlib.pyplot as plt
```

## Step 2. Declare the model variables

```python
K = 10 # Hydraulic conductivity [m/d]
delx = 5 #cell x size
dely = 1 #cell y size
delz = 5 #cell z size
nlay = 31 #number of layers
nrow = 1 #number of rows
ncol = 120 #number of cols
stress_periods = 1 # number of stress periods
length_time = 7300 # lengh of stress period (days)
ne = 0.1 # effective porosity
alpha_l = 0 # longitudinal dispersivity
diff = 0.308 #m2/d #Diffusion coefficient
Ss = 1e-4 # Specific Storage
salinity = 285.7 #kg/m3
buoyancy = 0.2 #Buoyancy
```

## Step 3. Create SEAWAT model and Packages with flopy framework

```python
modelname = "elder"

#Create the model:
swt = flopy.seawat.Seawat(modelname, exe_name=os.path.join('..','swt_v4x64'), model_ws=".")

#Define the top elevation
elder_top = 5

#Define the bottom elevation for each layer:
elder_botm = np.linspace(elder_top - delz, elder_top - nlay*delz, nlay)

# Add DIS package to the MODFLOW model
dis = flopy.modflow.ModflowDis(swt, nlay = nlay, nrow = nrow, ncol = ncol, nper=stress_periods, delr=delx,
                               delc=dely, laycbd=0, top=elder_top, steady = False,
                               botm=elder_botm, perlen=7200, nstp=720)


# Variables for the BAS package
ibound = np.ones((nlay, nrow, ncol), dtype=np.int32)

#Defining the constant head and inactive cells
ibound[0,:,0:30] = 0
ibound[0,:,90:120] = 0
ibound[1,:,0] = -1
ibound[1,:,119] = -1

#Definining the starting heads:
strt = np.ones((nlay, nrow, ncol), dtype = np.float32)*10
strt[ibound == -1] = 0

#Creating the Bas package:
bas = flopy.modflow.ModflowBas(swt, ibound, strt)
```

```python
#LPF package:
ipakcb = 53 #optional

#Making the LPF package:
lpf = flopy.modflow.ModflowLpf(swt, hk=K, vka=K, ss = Ss, laytyp=0, ipakcb = ipakcb)
```

```python
#Defining the PCG solver for the flow model:
pcg = flopy.modflow.ModflowPcg(swt, hclose=1.0e-8)
```

```python
#Defining the OC package for the flow Model

oc = flopy.modflow.ModflowOc(
    swt,
    stress_period_data={(0, 0): ["save head", "save budget"]},
    compact=True,
)
```

## Step 4. Create the MT3DMS model in the flopy framework

```python
#Create MT3DMS model structure

#Defininf the time intervals for exporting:
timprs = np.linspace(30,7200,30)

#Defining the initial concentration array:
initial_c = np.zeros((nlay,nrow,ncol))
initial_c[0,:,30:90] = 285.7

#Defininf the icbound array:
icbund = np.ones((nlay,nrow,ncol), np.float32)
icbund[0,:,30:90] = -1
icbund[nlay-1,:,:] = -1

#Creating the BTN package
btn = flopy.mt3d.Mt3dBtn(
    swt,
    nprs=30,
    timprs = timprs,
    prsity=ne,
    icbund = icbund,
    sconc=initial_c,
    ifmtcn=0,
    ncomp=1
)

#Advection package with mixelm = -1
adv = flopy.mt3d.Mt3dAdv(swt, mixelm=-1)

#Dispersion package:
dsp = flopy.mt3d.Mt3dDsp(swt, al=alpha_l, trpt=0.1, trpv=0.1, dmcoef=0.308)

#Defining GCG package:
gcg = flopy.mt3d.Mt3dGcg(swt, iter1=5000, mxiter=1, isolve=1, cclose=1e-7)

from flopy import mt3d
itype = mt3d.Mt3dSsm.itype_dict()
print(itype)

#Defining the stress variables for the SSM model:
ssm_data = {}
dat = []
for i in range(30,90):
    dat.append((0,0,i, 285.7, itype['CC']))

ssm_data[0] = dat

# Creating the SSM package:
ssm = flopy.mt3d.Mt3dSsm(swt, stress_period_data = ssm_data)
```

## Step 5. Defining the SEAWAT Variable Density flow package (VDF)

```python
vdf = flopy.seawat.SeawatVdf( # Parameters defined to imitate the VDF output from the exercise and PMWIN model
    swt,
    mt3drhoflg = -1,
    mfnadvfd = 1,
    nswtcpl = 0,
    iwtable=0,
    denseref = 1000,
    drhodprhd = 0,
    prhdref =0,
    nsrhoeos = 1,
    mtrhospec = 1,
    drhodc = 0.1,
    crhoref = 0,
    dt0 = 0.01
)
```

```python
swt.check() #Checking for model errors
```

## Step 6. Checking the initial concentration Array:

```python
xsect = flopy.plot.PlotCrossSection(swt, line= { "row": 0})
conc = xsect.plot_array(initial_c)
plt.colorbar(conc, shrink=0.5)
patches = xsect.plot_ibound()
```

## Step 7. Checking the Initial Head Array:

```python
xsect = flopy.plot.PlotCrossSection(swt, line= { "row": 0})
initial_head = xsect.plot_array(strt)
plt.colorbar(initial_head, shrink=0.5)
```

```python
swt.write_input()
success, buff = swt.run_model(silent=True, report=True)
if not success:
    raise Exception("SEAWAT did not terminate normally.")
```

## Final Concentration Cross-Section

```python
import flopy.utils.binaryfile as bf
ucnobj = bf.UcnFile("MT3D001.UCN", model=swt)
times = ucnobj.get_times()
print(len(times))
times_2 = times
len(times_2)
concentration = ucnobj.get_data(totim=times[-1])
```

```python
cbbobj = bf.CellBudgetFile("elder.cbc")
times = cbbobj.get_times()
print(times)
qx = cbbobj.get_data(text="flow right face", totim=times[-1])[0]
qy = np.zeros((nlay, nrow, int(ncol)), dtype=np.float32)
qz = cbbobj.get_data(text="flow lower face", totim=times[-1])[0]
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1, 1, 1)
pmv = flopy.plot.PlotCrossSection(model=swt, ax=ax, line={"row": 0})
arr = pmv.plot_array(concentration, masked_values = [1e30])
pmv.plot_vector(qx, qy, -qz, color="white", kstep=9, hstep=12)
plt.colorbar(arr, shrink=0.5, ax=ax, label = "Concentration")
ax.set_title("Simulated Concentrations");
```

## Breakthrough Curves - Observation Points

```python
points_coord = [(250,0.99,14),(300,0.99,14),(400,0.99,14)] #Coordinate of the points
points = []
for x,y,lay in points_coord:
    r, c = dis.get_rc_from_node_coordinates(x, y)
    point = (lay, r, c)
    points.append(point)

for point in points:
    ts = ucnobj.get_ts(point)
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(1, 1, 1)
    ttl = "Concentration at cell ({0},{1},{2})".format(point[0] + 1, point[1] + 1, point[2] + 1)
    ax.set_title(ttl)
    ax.set_xlabel("time [yrs]")
    ax.set_ylabel("Salt [g/L]")
    ax.plot(ts[:, 0]/365, ts[:, 1], "bo-")
```
